## Chart 10 — Key EDA Findings

1. **NAV Growth**: All 40 schemes showed consistent NAV appreciation from 2022–2026. The 2023 bull run pushed most funds to all-time highs, while the 2024 correction caused a temporary pullback before recovery. [Ref: Chart 1]

2. **SBI Dominance**: SBI Mutual Fund leads all fund houses by AUM with ₹12.5L Cr, nearly double the next competitor, indicating dominant market trust. [Ref: Chart 2]

3. **SIP All-Time High**: Monthly SIP inflows reached an all-time high of ₹31,002 Cr in Dec 2025, reflecting growing retail investor participation in mutual funds. [Ref: Chart 3]

4. **Category Seasonality**: Large Cap and Flexi Cap categories receive highest inflows consistently across all months, while Sectoral funds peak during specific months. [Ref: Chart 4]

5. **Young Investor Dominance**: The 26–35 age group contributes the highest SIP volume, showing that millennials are the primary driver of SIP growth in India. [Ref: Chart 5]

6. **Gender Split**: Male investors contribute approximately 60% of transactions while female investors account for 40%, with the gap narrowing in recent years. [Ref: Chart 5]

7. **Geographic Concentration**: Maharashtra, Delhi, and Karnataka are the top 3 states by SIP inflow. T30 cities dominate with over 70% of total investments. [Ref: Chart 6]

8. **Folio Growth**: Unique investor folios nearly doubled from Jan 2022 to Dec 2025, driven by digital onboarding, SIP awareness campaigns, and younger demographics. [Ref: Chart 7]

9. **High Correlation Among Large Caps**: Large Cap funds show correlation above 0.85 with each other, indicating they move closely with the market. Small Cap funds show lower correlation, offering better diversification. [Ref: Chart 8]

10. **Sector Concentration Risk**: Financial Services and IT together account for over 40% of equity fund portfolio weights, creating concentration risk that investors should be aware of. [Ref: Chart 9]
